## **Question 1: Difference between AI, ML, DL, and Data Science**
### **Artificial Intelligence (AI)**
AI is the broad field of creating machines that mimic human intelligence.
### **Machine Learning (ML)**
ML is a subset of AI where models learn patterns from data.
### **Deep Learning (DL)**
DL is a subset of ML using neural networks with multiple layers.
### **Data Science**
Data Science involves end‑to‑end data handling: collection, cleaning, analysis, ML, and visualization.

## **Question 2: Overfitting & Underfitting**
### **Overfitting**
Model learns noise; high train accuracy, low test accuracy.
### **Underfitting**
Model too simple; low train & test accuracy.
### **Preventing Both**
- Cross‑validation
- Regularization
- More data / features
- Model simplification or complexity adjustment.

## **Question 3: Handling Missing Values**
1. **Deletion** – remove missing rows.
2. **Mean/Median Imputation** – fill numeric values.
3. **Predictive Modeling** – predict missing entries using ML.

## **Question 4: Imbalanced Dataset**
When one class dominates the dataset.
- **SMOTE** – synthetic oversampling.
- **Random over/under sampling**.
- **Class weights** in models.

## **Question 5: Feature Scaling**
Scaling ensures fair contribution of all features.
- **Min‑Max Scaling** → (0–1)
- **Standardization** → mean 0, std 1.

## **Question 6: Label vs One‑Hot Encoding**
- **Label Encoding** for ordinal variables.
- **One‑Hot Encoding** for nominal variables.

## **Question 7: Google Play Store Dataset Analysis**

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/googleplaystore.csv"
df = pd.read_csv(url)

df = df[df['Rating'].notnull()]
df = df[df['Category'].notnull()]

category_ratings = df.groupby('Category')['Rating'].mean().sort_values(ascending=False)
category_ratings
```

```python
plt.figure(figsize=(12,6))
category_ratings.plot(kind='bar')
plt.title("Average Rating by App Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.show()
```

## **Question 9: Flight Price Prediction Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_price.csv"
df = pd.read_csv(url)

price_vs_days = df.groupby('days_left')['price'].mean()
price_vs_days
```

```python
plt.figure(figsize=(10,5))
plt.plot(price_vs_days.index, price_vs_days.values)
plt.title("Average Price vs Days Left")
plt.xlabel("Days Left")
plt.ylabel("Average Price")
plt.grid(True)
plt.show()
```

## **Question 10: HR Analytics Dataset**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/main/hr_analytics.csv"
df = pd.read_csv(url)

corr = df.corr(numeric_only=True)["Attrition"].sort_values(ascending=False)
print(corr)

sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm')
plt.show()
```
