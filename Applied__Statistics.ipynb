1. What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?

Type I error occurs when a true null hypothesis is wrongly rejected (false alarm), while Type II error happens when a false null hypothesis is not rejected (missed detection), and both impact decision-making by either leading to unnecessary actions or overlooking real effects.

2.What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?

The P-value is a measure of how compatible your observed data is with the assumption that the null hypothesis is true. In simple terms, it tells you the probability of seeing results as extreme (or more extreme) than what you observed, purely by chance, if the null hypothesis holds.

Small P-value (e.g., < 0.05): Strong evidence against the null hypothesis â†’ you may reject it.

Large P-value: Weak evidence against the null hypothesis â†’ you fail to reject it.

Interpretation nuance: A small P-value does not prove the alternative hypothesis; it only suggests the data is unlikely under the null.

4. What is a confidence interval, and how does the margin of error influence
its width and interpretation?

A confidence interval is a range of values, derived from sample data, that likely contains the true population parameter with a specified confidence level (e.g., 95%). The margin of error determines how wide the interval is: a larger margin of error produces a wider interval, reflecting greater uncertainty, while a smaller margin of error yields a narrower interval, indicating more precision. In interpretation, the interval shows the degree of reliability in estimatesâ€”narrow intervals suggest stronger evidence, wide intervals imply less certainty.

5. Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

ANOVA (Analysis of Variance) tests whether the means of three or more groups are significantly different. Its purpose is to extend hypothesis testing beyond two groups by comparing variance between groups to variance within groups. Key assumptions include normally distributed data, independent observations, and equal variances across groups. If the between-group variance is much larger than within-group variance, ANOVA suggests at least one group mean differs, helping identify differences across multiple categories simultaneously.

```python
#Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.


import numpy as np
from scipy.stats import norm


data = np.array([52, 55, 49, 50, 48, 53, 51])
sample_mean = np.mean(data)
population_mean = 50
population_sd = 5

z_score = (sample_mean - population_mean) / (population_sd / np.sqrt(len(data)))
p_value = 1 - norm.cdf(abs(z_score))

print("Z-score:", z_score)
print("P-value:", p_value)
```

```python
#: Write a Python function to calculate and visualize the confidence interval for a sample mean.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_ci(data, confidence=0.95):
    mean = np.mean(data)
    sd = np.std(data, ddof=1)
    n = len(data)

    z = norm.ppf((1 + confidence) / 2)
    moe = z * (sd / np.sqrt(n))

    lower = mean - moe
    upper = mean + moe

    print(f"{confidence*100}% Confidence Interval: ({lower}, {upper})")

    plt.errorbar(1, mean, yerr=moe, fmt='o')
    plt.title("Confidence Interval for Sample Mean")
    plt.xlim(0, 2)
    plt.show()

data = np.random.randn(30)
plot_ci(data)
```

```python
#Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

import numpy as np
from scipy.stats import chisquare

# Observed and expected frequencies
observed = np.array([25, 30, 20, 25])
expected = np.array([25, 25, 25, 25])

chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("P-value:", p_value)
```
